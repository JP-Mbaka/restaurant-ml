```python
import pandas as pd
```

```python
# load excel file

df = pd.read_excel('data.xlsx')
```

```python
df.head()
```

```python
# Rename Columns for ease
df.columns = ['time','age','gender','class','breakfast', 'lunch', 'dinner', 'dislikes','allergies','comments']
```

```python
df.head()
```

```python
# Create unified preference list
def get_pref_vector(row):
    foods = []
    for col in ['breakfast','lunch','dinner']:
        if pd.notnull(row[col]):
            foods +=[f.strip().lower() for f in row[col].split(',')]

    return foods
```

```python
df['preferences'] = df.apply(get_pref_vector,axis=1)
df['preferences'].head()
```

```python
df.head()
```

```python
#len(df.columns)
df.columns
```

```python
# Get all unique food items
all_foods = set(food for sublist in df['preferences'] for food in sublist)
all_foods
```

```python
# create binary columns for each food
for food in all_foods:
    df[food] = df['preferences'].apply(lambda x:1 if food in x else 0)

# user0food matrix
user_food_matrix = df[list(all_foods)]
user_food_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# compute Cosine Similarity between users
user_sim = cosine_similarity(user_food_matrix)
user_sim
```

```python
Similarity_df = pd.DataFrame(user_sim, index = df.index, columns = df.index)
Similarity_df
```

```python
# Function to clean comma-separated fields
def clean_food_list(cell):
    if pd.isna(cell):
        return []
    return [f.strip().lower() for f in cell.split(',')]

#choose target user (e.g. index 1)
target_user = 1

# Top 3 Similar users(excluding self)
similar_users = Similarity_df[target_user].drop(target_user).sort_values(ascending=False)
similar_users
```

```python
# mean_prefence scores from similar uses
mean_scores =  user_food_matrix.iloc[similar_users].mean()
mean_scores
```

```python
# Remove foods the user already likes/dislikes & allergies
liked = set(df.loc[target_user,'preferences'])
disliked = set(clean_food_list(df.loc[target_user,'dislikes']))
allergic = set(clean_food_list(df.loc[target_user,'allergies']))
blocked = liked.union(disliked).union(allergic)
```

```python
# Recommend top 10 new foods
recommendations = mean_scores.drop(labels=blocked,errors='ignore').sort_values(ascending=False).head(10)
print("Top 10 Recommended Foods:\n",recommendations)
```
